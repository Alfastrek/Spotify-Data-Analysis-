```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_tracks=pd.read_csv(r'D:\downloads\jupyter pbl\archive\SpotifyFeatures.csv')
df_tracks.head()
```

```python
df_least_popular_sorted=df_tracks.sort_values('popularity', ascending= True).head(10)
df_least_popular_sorted
```

```python
most_popular=df_tracks.query('popularity > 90', inplace= False).sort_values('popularity', ascending= False)
most_popular[:10]
```

```python
corr_df=df_tracks.drop(['key','mode'], axis=1).corr(method='pearson')
plt.figure(figsize =[14,6])
heatmap=sns.heatmap(corr_df, annot=True, fmt= '.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linecolor= 'Black')
heatmap.set_title('correlation heatmap between variable')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
plt.figure(figsize=[10,6])
sns.regplot(data=sample_df, y='loudness', x='energy', color= 'c')
```

```python
plt.figure(figsize=[10,6])
sns.regplot(data=sample_df, y= 'popularity', x ='acousticness', color= 'g')
```

```python
plt.figure(figsize=[15,6])
plt.title('Duration of the song in different genres')
sns.color_palette('rocket', as_cmap= True)
sns.barplot(y= 'genre', x = 'duration_ms', data = df_tracks)
```

```python
sns.set_style(style = 'darkgrid')
plt.figure(figsize=[10,6])
most_popular_genres = df_tracks.sort_values('popularity', ascending= False). head(10)
sns.barplot(y='genre', x='popularity', data= most_popular_genres).set(title= 'Top 10 genre by popularity')
```
